```python
import random

def calculate_nim_sum(heaps):
    nim_sum = 0
    for heap in heaps:
        nim_sum ^= heap
    return nim_sum

def ai_move(heaps):
    nim_sum = calculate_nim_sum(heaps)

    # If nim sum is 0, AI will make a random move
    for i in range(len(heaps)):
        target = heaps[i] ^ nim_sum
        if target < heaps[i]:
            remove = heaps[i] - target
            return i, remove
    # Random move if no winning move
    non_empty = [i for i in range(len(heaps)) if heaps[i] > 0]
    heap = random.choice(non_empty)
    return heap, 1

def display_heaps(heaps):
    print("\nCurrent Heaps:")
    for i, heap in enumerate(heaps):
        print(f"Heap {i+1}: {'O ' * heap}({heap})")

def nim_game():
    heaps = [3, 4, 5]  # You can change this to any heap configuration
    print(" Welcome to the Nim Game!")
    print(" You and the AI will take turns removing objects from a single heap.")
    print(" On your turn, choose a heap and how many to remove (at least 1).")
    print(" The player who removes the LAST object loses!")

    player_turn = True  # You go first

    while any(heap > 0 for heap in heaps):
        display_heaps(heaps)

        if player_turn:
            try:
                heap = int(input("Choose a heap (1, 2, or 3): ")) - 1
                if heaps[heap] == 0:
                    print("âš ï¸ This heap is empty. Try again.")
                    continue
                count = int(input(f"How many to remove from heap {heap+1}? "))
                if count < 1 or count > heaps[heap]:
                    print("âš ï¸ Invalid number of objects. Try again.")
                    continue
                heaps[heap] -= count
            except:
                print("âŒ Invalid input. Try again.")
                continue
        else:
            print("\nğŸ’» AI is making a move...")
            heap, count = ai_move(heaps)
            heaps[heap] -= count
            print(f"ğŸ¤– AI removed {count} from Heap {heap+1}")

        if sum(heaps) == 0:
            display_heaps(heaps)
            if player_turn:
                print("\nâŒ You took the last object. You lose!")
            else:
                print("\nğŸ‰ AI took the last object. You win!")
            break

        player_turn = not player_turn

# Run the game
nim_game()

```
